```python
def all_in_one(url1, url2, url3, url4, url5):
    X = 18
    Y = 4
    url_df = url1
    df1 = pd.read_html(url_df)[1]
    df1 = df1.iloc[X:]
    df1 = df1.reset_index(drop=True)
    
    url_df = url2
    df2 = pd.read_html(url_df)[1]
    
    url_df = url3
    df3 = pd.read_html(url_df)[1]
    
    url_df = url4
    df4 = pd.read_html(url_df)[1]
    
    url_df = url5
    df5 = pd.read_html(url_df)[1]
    df5 = df5.iloc[:Y]
    df5 = df5.reset_index(drop=True)
    
    df = pd.concat([df1, df2, df3, df4, df5], axis=0)

    # Reset the index of the concatenated DataFrame
    df = df.reset_index(drop=True)

    # Display the concatenated DataFrame
    df['xGd'] = df['xG'] - df['xGA']
    df = df[['Date', 'xG', 'xGA', 'xGd']]
    
    window_size = 11

    # Calculate the rolling average for the entire column
    df['SMA'] = df['xGd'].rolling(window=window_size, min_periods=1).mean()

    # Leave the first 10 values in the RollingAverage column as they are
    df['SMA'].iloc[:window_size] = df['xGd'].iloc[:window_size]
    
    return df
    
    
```

```python
import matplotlib.pyplot as plt

def plot_team_comparison(team_data_list, title, team_names):
    # Define a dictionary to map team names to fixed colors
    team_colors = {
        "Arsenal": 'red',
        "City": '#98C5E9',
        "United": 'black',
        "Chelsea": 'blue',
        "Spurs": 'white',
        "Liverpool": 'pink'
        # Add more teams and colors as needed
    }

    fig = plt.figure(figsize=(7, 4), dpi=300)
    ax = plt.subplot(111)
    ax.patch.set_color('#4a4e69')
    fig.set_facecolor('#4a4e69')

    ax.spines[["top", "right"]].set_visible(False)
    ax.spines[["left", "bottom"]].set_color("grey")

    for i, team_data in enumerate(team_data_list):
        # Plot the specified team's data using the fixed color
        team_name = team_names[i]
        color = team_colors.get(team_name, 'red')  # Default to 'red' if not found
        ax.plot(team_data.matches, team_data.SMA, zorder=4, label=team_name, lw=1.5, color=color)

    plt.ylabel('xGd SMA', fontsize=10, color='w')
    plt.yticks(fontsize=8, color='w')
    ax.set_ylim(-0.75, 2)

    ax.set_xticklabels([])
    ax.tick_params(axis='x', which='both', bottom=False)

    # Add a line to mark the season start
    ax.plot(
        [37, 37],  # Adjust this line as needed
        [ax.get_ylim()[0], ax.get_ylim()[1]],
        ls=":",
        lw=0.6,
        color="grey",
        zorder=2
    )
    
    # Add a line to mark 21/22 begin
    ax.plot(
        [37,37], # 38 games per season
        [ax.get_ylim()[0], ax.get_ylim()[1]],
        ls = ":",
        lw = 0.6,
        color = "grey",
        zorder = 2
        )

    # Add a line to mark 22/23 begin
    ax.plot(
        [75,75], # 38 games per season
        [ax.get_ylim()[0], ax.get_ylim()[1]],
        ls = ":",
        lw = 0.6,
        color = "grey",
        zorder = 2
        )

    # Add a line to mark 23/24 begin
    ax.plot(
        [113,113], # 38 games per season
        [ax.get_ylim()[0], ax.get_ylim()[1]],
        ls = ":",
        lw = 0.6,
        color = "grey",
        zorder = 2
    )


    plt.axhline(y = 2, color = 'grey', linestyle = '-')

    ax.annotate(
        xy = (112, -0.6),
        xytext = (20, 10),
        textcoords = "offset points",
        text = "23/24 start",
        size = 6,
        color = "white",
        alpha=0.7,
        arrowprops=dict(
            arrowstyle="->", shrinkA=0, shrinkB=5, color="w", linewidth=0.85,
            connectionstyle="angle3,angleA=50,angleB=-30"
       )#  Arrow to connect annotation
    )

    ax.annotate(
        xy = (74, -0.6),
        xytext = (20, 10),
        textcoords = "offset points",
        text = "22/23 start",
        size = 6,
        color = "white",
        alpha=0.7,
        arrowprops=dict(
            arrowstyle="->", shrinkA=0, shrinkB=5, color="w", linewidth=0.85,
            connectionstyle="angle3,angleA=50,angleB=-30"
       )#  Arrow to connect annotation
    )

    ax.annotate(
        xy = (36, -0.6),
        xytext = (20, 10),
        textcoords = "offset points",
        text = "21/22 start",
        size = 6,
        color = "white",
        alpha=0.7,
        arrowprops=dict(
            arrowstyle="->", shrinkA=0, shrinkB=5, color="w", linewidth=0.85,
            connectionstyle="angle3,angleA=50,angleB=-30"
       )#  Arrow to connect annotation
    )

    # Customize title
    ax.text(
        -4,  # X-coordinate of the text
        2.1,  # Y-coordinate of the text (adjust as needed)
        title,  # Custom title
        fontsize=12,
        color="white",
        weight='bold',
        fontfamily='serif',
        horizontalalignment='left'
    )

    # Customize label position based on team_data
    for i, team_data in enumerate(team_data_list):
        x_end_team = team_data.matches.iloc[-1]
        y_end_team = team_data.SMA.iloc[-1] - 0.05
        team_name = team_names[i]
        color = team_colors.get(team_name, 'red')
        ax.text(x_end_team, y_end_team, team_names[i], fontsize=8, color=color, va='center')

    # Add data credit
    ax.text(
        125,  # X-coordinate of the text (adjust as needed)
        -0.9,  # Y-coordinate of the text (adjust as needed)
        "Data via fbref",  # Text label
        fontsize=8,
        color="white",
        alpha=0.5,
        style='italic',  # Set to 'italic' for italics
        horizontalalignment='right'
    )
    ax.text(
        30,  # X-coordinate of the text (adjust as needed)
        -0.9,  # Y-coordinate of the text (adjust as needed)
        "Data from boxing day 2019",  # Text label
        fontsize=8,
        color="white",
        alpha=0.5,
        style='italic',  # Set to 'italic' for italics
        horizontalalignment='right'
    )

    plt.tight_layout()
    plt.show()

```
